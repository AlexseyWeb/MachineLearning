```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
values = [1, 2, 3, 4, 5, 6, 7, 8, 9]
squares = [x**2 for x in values]
```

```python
plt.plot(values, squares)
plt.show()
```

```python
plt.scatter(values, squares)
plt.show()
```

```python
values = list(range(100))
pows = [pow(value, 3) for value in values]
plt.scatter(values, pows)
plt.show()
```

```python
plt.scatter(values, pows, c=pows, cmap=plt.cm.Blues, edgecolors=None, s=10)
plt.show()
```

```python
values = list(range(0, 50))
squares = [pow(value, 2) for value in values]
cubes = [pow(value, 3) for value in values]

fig, (ax, ax2) = plt.subplots(2, 1, sharex=True)
ax.scatter(values, squares)
ax.set_title("Square numbers")
ax2.plot(values, cubes)
ax2.set_title("Cube numbers")
plt.show()
```

```python
values = list(range(0, 50))
squares = [pow(value, 2) for value in values]
cubes = [pow(value, 3) for value in values]

fig, (ax, ax2) = plt.subplots(1, 2, sharey=True)
ax.scatter(values, squares)
ax.set_title("Square numbers")
ax2.plot(values, cubes)
ax2.set_title("Cube numbers")
plt.show()
```

```python
data = {
    'Year': [2020, 2021, 2022, 2023, 2024],
    'Sales': [100, 150, 130, 180, 270]
}
df = pd.DataFrame(data)
df.plot(x='Year', y='Sales', kind='line')
plt.title("Years sales")
plt.show()
```

```python
data = {
    'Year': [2020, 2021, 2022, 2023, 2024],
    'Sales': [100, 150, 130, 180, 270]
}
df = pd.DataFrame(data)
df.plot(y='Sales', kind='pie', labels=df['Year'])
plt.title("Years sales")
plt.show()
```

```python
data = {'Category': ['First', 'First', 'Second', 'Second', 'Third', 'Third'],
        'Value': [10, 20, 25, 30, 50, 60],
       }
df = pd.DataFrame(data)
df.head()
```

```python
df.boxplot(by='Category', column='Value')
plt.title('Values distributes by category')
plt.show()
```

```python
data = {
        'Month': ['January', 'Febrary', 'March', 'April', 'May'],
        'Week 1 Temp': [6, 10, 12, 14, 15],
        'Week 2 Temp': [16, 18, 19, 20, 17],
        'Week 3 Temp': [20, 21, 22, 20, 21],
        'Week 4 Temp': [23, 24, 22, 20, 25],
       }
df = pd.DataFrame(data)
df.set_index('Month', inplace=True)
```

```python
df.head()
```

```python
sns.heatmap(df, cmap='coolwarm', annot=True)
plt.title("Weekly Templature ")
```
